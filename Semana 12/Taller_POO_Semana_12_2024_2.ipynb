```python
import pandas as pd
df = pd.read_csv("tita.csv").set_index("PassengerId")
print("Dimensiones (fila, col): ", df.shape)
print("Cantidad de datos que contiene: ", df.size)
print("Nombre Columnas: ", df.columns)
print("Nombre filas: ", df.index)
print("Tipos de datos: ", df.dtypes)
```

```python
df.head(2)
```

```python
df.tail(5)
```

```python
df.loc[148]
```

```python
print("Filas pares: ")
df.iloc[1::2]
```

```python
print("Nombres ordenados: ", df[df['Pclass'] == 1]['Name'].sort_values())
print("Porcentaje personas Sobrevivieron y murieron: ", df['Survived'].value_counts(normalize=True)*100)
```

```python
print("Eliminar los que tienen edad desconocida: ")
df= df.dropna(subset=['Age'])
df
```

```python
print("Edad promedio de mujeres por clase: ", df[df['Sex'] == 'female'].groupby('Pclass')['Age'].mean())
df['MenorEdad'] = df['Age'] < 18
df
```

```python
print("Porcentaje menores mayores edad: ", df.groupby(['Pclass', 'MenorEdad'])['Survived'].mean()*100)
```
