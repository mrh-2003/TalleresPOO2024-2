```python
def factorial(num):
  fac = 1
  for i in range(1, num + 1):
    fac*= i
  return fac
def suma(n):
  sum = 0
  for i in range(1, n + 1):
    sum += i**2/(2*factorial(i))
  return sum

n = -1
while not(n >= 1 and n <= 120):
  try:
    n = int(input("Ingrese n: "))
  except ValueError:
    print("Ingresa datos correctos, no seas gil")
print("La sumatoria es:", suma(n))
```

```python
def analizar_nota(nota):
  if nota >= 9:
    print("Sobresaliente")
  elif nota >= 7:
    print("Notable")
  elif nota == 6:
    print("Bien")
  elif nota == 5:
    print("Suficiente")
  else:
    print("?")

nota = -1
while not(nota >= 0 and nota <= 10):
  try:
    nota = int(input("Ingrese nota: "))
  except ValueError:
    print("Ingresa datos correctos, por eso eres bica")
analizar_nota(nota)
```

```python
def suma(n, b):
  sum = 0
  cont = 4
  for i in range(1, n + 1):
    sum += (-1) ** (i + 1) * i * 2 * b / cont
    cont += 3
  print("La sumatoria es: ", sum)
  print("La sumatoria es: ", round(sum, 3) )
  print(f"La sumatoria es: {sum:.3f}")


n = -1
while not(n >= 1):
  try:
    n = int(input("Ingrese n: "))
    if not(n >= 1):
      print("Ingresa solo enteros positivos")
  except ValueError:
    print("Ingresa solo enteros")
b = -1
while not(b >= 1 and b <= 5):
  try:
    b = float(input("Ingrese b: "))
    if not(b >= 1 and b <= 5):
      print("Ingresa un real en el intervalo correcto")
  except ValueError:
    print("Ingresa solo numeros")
suma(n, b)
```
