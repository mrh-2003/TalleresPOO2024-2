```python
def menu():
  op = -1
  while not(1<= op <= 4):
    print("(1) Aadir cliente\n(2) Eliminar cliente\n(3) Listar todos los clientes\n(4) Terminar.")
    try:
      op = int(input("Ingrese la opcion: "))
    except ValueError:
      print("La opcion es un numero entero")
  return op

def existe_dni(dnis, dni):
  return dnis.count(dni) == 1

op = 0
clientes = {}
while op != 4:
  op = menu()
  if op == 1:
    dni = ''
    while not(dni.isnumeric() and len(dni) == 8 and not(existe_dni(list(clientes.keys()), dni))):
      dni = input('Ingrese dni: ')
    nombre = ''
    while not(len(nombre) != 0 and nombre.isalpha()):
       nombre = input('Ingrese nombre: ')
    apellido = ''
    while not(len(apellido) != 0 and apellido.isalpha()):
       apellido = input('Ingrese apellido: ')
    telefono = ''
    while not(telefono.isnumeric() and len(telefono) == 9):
      telefono = input('Ingrese telefono: ')
    correo = nombre[0].lower() + apellido + "@empsac.com"
    preferente = -1
    while not(preferente == 1 or preferente == 2):
      try:
        preferente = int(input("Ingrese preferente: "))
      except ValueError:
        print("Debes ingresar un enetero")
    if preferente == 1:
      pref = True
    else:
      pref = False
    clientes[dni] = {
        "nombre": nombre,
        "apellido": apellido,
        "telefono": telefono,
        "correo": correo,
        "preferente": pref,
    }
  elif op == 2:
    dni = ''
    while not(dni.isnumeric() and len(dni) == 8 and existe_dni(list(clientes.keys()), dni)):
      dni = input('Ingrese dni: ')
    clientes.pop(dni)
  elif op == 3:
    for key, value in clientes.items():
      print(key, value["nombre"], value["telefono"], value["correo"], value["apellido"], value["preferente"])
```
