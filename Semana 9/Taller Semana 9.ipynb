```python
class Encomienda:
  def __init__(self, codigo, remitente, destinatario, direccion, pesoKg, volumenM3, costoKg, costoM3):
    self.__codigo = codigo
    self.__remitente = remitente
    self.__destinatario = destinatario
    self.__direccion =direccion
    self.__pesoKg =pesoKg
    self.__volumenM3 =volumenM3
    self.__costoKg =costoKg
    self.__costoM3 = costoM3
    self.__encomienda={
        " Cdigo": self.__codigo,
        " Remitente": self.__remitente,
        " Destinatario": self.__destinatario,
        " Direccin de Entrega": self.__direccion,
        " Peso en Kg": self.__pesoKg,
        " Volumen en m3": self.__volumenM3,
        " Costo por Kg": self.__costoKg,
        " Costo por m3": self.__costoM3
    }

  def get_costo_por_peso(self):
    return self.__costoKg * self.__pesoKg
  def get_costo_por_volumen(self):
    return self.__costoM3 * self.__volumenM3
  def get_costo_por_envio(self):
    return self.get_costo_por_volumen() + self.get_costo_por_peso()

  def get_enconmienda(self):
    return self.__encomienda
  def get_codigo(self):
    return self.__codigo
  def get_remitente(self):
    return self.__remitente
  def get_destinatario(self):
    return self.__destinatario
  def get_direccion(self):
    return self.__direccion
  def get_pesoKg(self):
    return self.__pesoKg
  def get_volumenM3(self):
    return self.__volumenM3
  def get_costoKg(self):
    return self.__costoKg
  def get_costoM3(self):
    return self.__costoM3

  def set_codigo(self, codigo):
    self.__codigo = codigo
  def set_remitente(self, remitente):
    self.__remitente = remitente
  def set_destinatario(self, destinatario):
    self.__destinatario = destinatario
  def set_direccion(self, direccion):
    self.__direccion = direccion
  def set_pesoKg(self, pesoKg):
    self.__pesoKg = pesoKg
  def set_volumenM3(self, volumenM3):
    self.__volumenM3 = volumenM3
  def set_costoKg(self, costoKg):
    self.__costoKg = costoKg
  def set_costoM3(self, costoM3):
    self.__costoM3 = costoM3


enc1 = Encomienda(123, "Matheo", "Milene", "Lima", 15, 80, 10, 30)
enc2 = Encomienda(123, "Matheo", "Ana", "Tacna", 45, 10, 15, 450)
enc3 = Encomienda(123, "Matheo", "Jazmin", "Ica", 32, 50, 14, 90)

print("Encomienda 1: ")
print(enc1.get_enconmienda())
print("El costo por peso es: ", enc1.get_costo_por_peso())
print("El costo por volumen es: ", enc1.get_costo_por_volumen())
print("El costo por envio es: ", enc1.get_costo_por_envio())
print("Encomienda 2: ")
print(enc2.get_enconmienda())
print("El costo por peso es: ", enc2.get_costo_por_peso())
print("El costo por volumen es: ", enc2.get_costo_por_volumen())
print("El costo por envio es: ", enc2.get_costo_por_envio())
print("Encomienda 3: ")
print(enc3.get_enconmienda())
print("El costo por peso es: ", enc3.get_costo_por_peso())
print("El costo por volumen es: ", enc3.get_costo_por_volumen())
print("El costo por envio es: ", enc3.get_costo_por_envio())
```

```python
import random as r

class Terreno:
  def __init__(self, codigo, ubicacion, area, valor_comercial, valor_predial):
    self.__codigo =codigo
    self.__ubicacion = ubicacion
    self.__area = area
    self.__valor_comercial = valor_comercial
    self.__valor_predial = valor_predial

    self.__terreno = {
        " Cdigo": self.__codigo,
        " Ubicacin": self.__ubicacion,
        " rea": self.__area,
        " Valor m2 Comercial": self.__valor_comercial,
        " Valor m2 Predial.":self.__valor_predial
    }

  def get_valor_predial(self):
    return self.__area * self.__valor_predial
  def get_valor_comercial(self):
    return self.__area * self.__valor_comercial
  def get_margen_ganancia(self):
    return self.get_valor_comercial() - self.get_valor_predial()
  def get_terreno(self):
    return self.__terreno
  def get_codigo(self):
    return self.__codigo
  def get_ubicacion(self):
    return self.__ubicacion
  def get_area(self):
    return self.__area
  def get_valor_comercial(self):
    return self.__valor_comercial
  def get_valor_predial(self):
    return self.__valor_predial

  def set_codigo(self, codigo):
    self.__codigo = codigo
  def set_ubicacion(self, ubicacion):
    self.__ubicacion = ubicacion
  def set_area(self, area):
    self.__area = area
  def set_valor_comercial(self, valor_comercial):
    self.__valor_comercial = valor_comercial
  def set_valor_predial(self, valor_predial):
    self.__valor_predial = valor_predial

n = 10
terrenos = []
ubicaciones = ['Amazonas',
 'Ancash',
 'Apurimac',
 'Arequipa',
 'Ayacucho',
 'Cajamarca',
 'Callao',
 'Cusco',
 'Huancavelica',
 'Huanuco',
 'Ica',
 'Junn',
 'La Libertad',
 'Lambayeque',
 'Lima',
 'Loreto',
 'Madre de Dios',
 'Moquegua',
 'Pasco',
 'Piura',
 'Puno',
 'San Martn',
 'Tacna',
 'Tumbes',
 'Ucayali']

for i in range(n):
  terrenos.append(Terreno("P" + str(i), r.choice(ubicaciones), r.randint(50, 800), r.randint(200, 2500), r.randint(300, 1500)))

for terreno in terrenos:
  print("Terreno: ", terreno.get_terreno())
  print("Valor predial: ", terreno.get_valor_predial())
  print("Valor comercial: ", terreno.get_valor_comercial())
  print("Valor margen: ", terreno.get_margen_ganancia())
```
