```python
import random as r

def ingresar_n():
  n = -1
  while not(n >= 1 and n<= 40):
    try:
      n = int(input("Ingrese n: "))
      if not(n >= 1 and n<= 40):
        print("Ingresa un numero entero positivo")
    except ValueError:
      print("Ingresa un numero entero positivo")
  return n;

def generar():
  n = ingresar_n()
  lista = []
  for i in range(n):
      lista.append(r.randint(1, 9))
  return lista

def mostrar(lista):
  for num in lista:
    print(num, end= ' ')
  print()
def contar_repeticiones(lista):
  for num in set(lista):
    print(f"Cantidad de {num} es {lista.count(num)}")

def es_primo(num):
  cont = 0
  for i in range(1, num+1):
    if num % i == 0:
      cont+=1
  return cont == 2

def incrementar_primos(lista):
  for i in range(len(lista)):
    if es_primo(lista[i]):
      lista[i] += 1
lista = generar()
mostrar(lista)
contar_repeticiones(lista)
incrementar_primos(lista)
print("Ingremento de valores primos: ")
mostrar(lista)
```

```python
nombres = []
apellidos = []
codigos = []

def existe_codigo(codigos, codigo):
    return codigos.count(codigo) == 0  # si no existe True, si existe False

def ordenar(codigos, nombres, apellidos):
    auxCod = 0
    auxNom = ''
    auxApe = ''
    for i in range(0, len(codigos) - 1):
        for j in range(i + 1, len(codigos)):
            if codigos[i] > codigos[j]:
                auxCod = codigos[i]
                codigos[i] = codigos[j]
                codigos[j] = auxCod
                auxNom = nombres[i]
                nombres[i] = nombres[j]
                nombres[j] = auxNom
                auxApe = apellidos[i]
                apellidos[i] = apellidos[j]
                apellidos[j] = auxApe
            elif codigos[i] == codigos[j] and apellidos[i] > apellidos[j]:
                auxApe = apellidos[i]
                apellidos[i] = apellidos[j]
                apellidos[j] = auxApe
                auxNom = nombres[i]
                nombres[i] = nombres[j]
                nombres[j] = auxNom

op = 'S'
while op == 'S':
    codigo = ''
    while True:
        codigo = int(input("Ingrese codigo: "))
        if len(str(codigo)) == 8 and existe_codigo(codigos, codigo):
            break;
    codigos.append(codigo)

    nombre = ''
    while not(nombre != ''):
        nombre = input("Ingrese nombre: ")
    nombres.append(nombre)

    apellido = ''
    while not(apellido != ''):
        apellido = input("Ingrese apellido: ")
    apellidos.append(apellido)

    op = input('Quieres volver a repetir?: ').upper()

print('Cantidad de matriculados: ', len(codigos))
print(codigos)
print(nombres)
print(apellidos)

ordenar(codigos, nombres, apellidos)

print("Lista ordenada por cdigo y luego por apellido:")
print(codigos)
print(nombres)
print(apellidos)
```

```python
tarifa = float(input("Ingrese la tarifa por segundo (Soles/seg): "))
num_llamadas = int(input("Ingrese el nmero de llamadas a contabilizar: "))
total_segundos = 0
for i in range(num_llamadas):
    horas = int(input("Ingrese la duracin en horas: "))
    minutos = int(input("Ingrese la duracin en minutos: "))
    segundos = int(input("Ingrese la duracin en segundos: "))
    total_segundos += horas * 3600 + minutos * 60 + segundos
monto_a_facturar = total_segundos * tarifa
print(f"El monto a facturar es: S/ {monto_a_facturar:.2f}")
```
